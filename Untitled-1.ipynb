```python
# Filter rows where 'event_type' matches your list
#df is  processed_match_events in database
pass_df = df[df['Action'].isin(['Pass', 'Through Ball'])]
```

```python
import numpy as np

# Define success condition to keep the code clean
success_outcomes = ['Successful', 'Key Pass', 'Assist']

# 1. Base Passes (Total, Complete, Incomplete)
total_pass = pass_df['Act Team'].value_counts().reset_index(name='Total Pass')
complete = pass_df[pass_df['Outcome'].isin(success_outcomes)]['Act Team'].value_counts().reset_index(name='Complete Pass')
incomplete = pass_df[pass_df['Outcome'].isin(['Missed', 'Intercepted', 'Off-Side'])]['Act Team'].value_counts().reset_index(name='Incomplete Pass')

# 2. Own Half (Total & Successful)
own_half_tot = pass_df[pass_df['Start X'] <= 60]['Act Team'].value_counts().reset_index(name='Total Own Half')
own_half_succ = pass_df[(pass_df['Start X'] <= 60) & pass_df['Outcome'].isin(success_outcomes)]['Act Team'].value_counts().reset_index(name='Successful Own Half')

# 3. Opp Half (Total & Successful)
opp_half_tot = pass_df[pass_df['Start X'] >= 60]['Act Team'].value_counts().reset_index(name='Total Opp Half')
opp_half_succ = pass_df[(pass_df['Start X'] >= 60) & pass_df['Outcome'].isin(success_outcomes)]['Act Team'].value_counts().reset_index(name='Successful Opp Half')

# 4. Throw-ins (Total & Successful)
throw_tot = pass_df[pass_df['Type'] == 'Throw-in']['Act Team'].value_counts().reset_index(name='Total Throw-ins')
throw_succ = pass_df[(pass_df['Type'] == 'Throw-in') & pass_df['Outcome'].isin(success_outcomes)]['Act Team'].value_counts().reset_index(name='Successful Throw-ins')

# 5. Corners (Total & Successful - using original df)
corner_tot = df[df['Type'] == 'Corner Kick']['Act Team'].value_counts().reset_index(name='Total Corners')
corner_succ = df[(df['Type'] == 'Corner Kick') & df['Outcome'].isin(success_outcomes)]['Act Team'].value_counts().reset_index(name='Successful Corners')

# 6. Long Balls (Calculate Distance -> Total & Successful)
pass_df['pass_distance'] = np.sqrt((pass_df['End X'] - pass_df['Start X'])**2 + (pass_df['End Y'] - pass_df['Start Y'])**2)
long_tot = pass_df[pass_df['pass_distance'] >= 32]['Act Team'].value_counts().reset_index(name='Total Long Balls')
long_succ = pass_df[(pass_df['pass_distance'] >= 32) & pass_df['Outcome'].isin(success_outcomes)]['Act Team'].value_counts().reset_index(name='Successful Long Balls')

# 7. Passes IN Attacking 3rd (Start X >= 80 AND End X >= 80)
in_att_3rd_tot = pass_df[(pass_df['Start X'] >= 80) & (pass_df['End X'] >= 80)]['Act Team'].value_counts().reset_index(name='Total In Att 3rd')
in_att_3rd_succ = pass_df[(pass_df['Start X'] >= 80) & (pass_df['End X'] >= 80) & pass_df['Outcome'].isin(success_outcomes)]['Act Team'].value_counts().reset_index(name='Successful In Att 3rd')

# 8. Passes INTO Attacking 3rd (Start X < 80 AND End X >= 80)
into_att_3rd_tot = pass_df[(pass_df['Start X'] < 80) & (pass_df['End X'] >= 80)]['Act Team'].value_counts().reset_index(name='Total Into Att 3rd')
into_att_3rd_succ = pass_df[(pass_df['Start X'] < 80) & (pass_df['End X'] >= 80) & pass_df['Outcome'].isin(success_outcomes)]['Act Team'].value_counts().reset_index(name='Successful Into Att 3rd')

# 9. Crosses (Based on adjusted 120x80 pitch coordinates)
cross_df = pass_df[
    (pass_df['Start X'] >= 80) &                                    
    ((pass_df['Start Y'] <= 12) | (pass_df['Start Y'] >= 68)) &     
    (pass_df['End X'] >= 104) &                                     
    (pass_df['End Y'].between(20, 60)) &                            
    (pass_df['Type'] != 'Throw-in')                                 
]
cross_tot = cross_df['Act Team'].value_counts().reset_index(name='Total Crosses')
cross_succ = cross_df[cross_df['Outcome'].isin(success_outcomes)]['Act Team'].value_counts().reset_index(name='Successful Crosses')

# 10. Specific Incomplete Pass Types
missed = pass_df[pass_df['Outcome'] == 'Missed']['Act Team'].value_counts().reset_index(name='Missed Passes')
intercepted = pass_df[pass_df['Outcome'] == 'Intercepted']['Act Team'].value_counts().reset_index(name='Intercepted Passes')
offside = pass_df[pass_df['Outcome'] == 'Off-Side']['Act Team'].value_counts().reset_index(name='Off-Side Passes')

# 11. Progressive Passes (Total & Opponent Half)
prog_tot = pass_df[(pass_df['Start X'] < pass_df['End X']) & (pass_df['End X'] - pass_df['Start X'] > 15)]['Act Team'].value_counts().reset_index(name='Total Prog Passes')
prog_succ = pass_df[(pass_df['Start X'] < pass_df['End X']) & (pass_df['End X'] - pass_df['Start X'] > 15) & (pass_df['Outcome'].isin(success_outcomes))]['Act Team'].value_counts().reset_index(name='Successful Prog Passes')

prog_opp_tot = pass_df[(pass_df['Start X'] >= 60) & (pass_df['Start X'] < pass_df['End X']) & (pass_df['End X'] - pass_df['Start X'] > 15)]['Act Team'].value_counts().reset_index(name='Total Prog Opp Half')
prog_opp_succ = pass_df[(pass_df['Start X'] >= 60) & (pass_df['Start X'] < pass_df['End X']) & (pass_df['End X'] - pass_df['Start X'] > 15) & (pass_df['Outcome'].isin(success_outcomes))]['Act Team'].value_counts().reset_index(name='Successful Prog Opp Half')



#12. Ends inside the box
into_box_df = pass_df[
    ((pass_df['Start X'] < 102) | (~pass_df['Start Y'].between(20, 60))) &  
    (pass_df['End X'] >= 102) & (pass_df['End Y'].between(20, 60))          
]

# Get total and successful counts
into_box_tot = into_box_df['Act Team'].value_counts().reset_index(name='Total Into Box')
into_box_succ = into_box_df[into_box_df['Outcome'].isin(success_outcomes)]['Act Team'].value_counts().reset_index(name='Successful Into Box')

#13. Starts inside the box AND Ends inside the box
inside_box_df = pass_df[
    (pass_df['Start X'] >= 102) & (pass_df['Start Y'].between(20, 60)) &    
    (pass_df['End X'] >= 102) & (pass_df['End Y'].between(20, 60))          
]

# Get total and successful counts
inside_box_tot = inside_box_df['Act Team'].value_counts().reset_index(name='Total Inside Box')
inside_box_succ = inside_box_df[inside_box_df['Outcome'].isin(success_outcomes)]['Act Team'].value_counts().reset_index(name='Successful Inside Box')

# 14. Chain all merges together based on 'Act Team'
pass_stats = (total_pass
              .merge(complete, on='Act Team', how='left')
              .merge(incomplete, on='Act Team', how='left')
              .merge(own_half_tot, on='Act Team', how='left')
              .merge(own_half_succ, on='Act Team', how='left')
              .merge(opp_half_tot, on='Act Team', how='left')
              .merge(opp_half_succ, on='Act Team', how='left')
              .merge(throw_tot, on='Act Team', how='left')
              .merge(throw_succ, on='Act Team', how='left')
              .merge(corner_tot, on='Act Team', how='left')
              .merge(corner_succ, on='Act Team', how='left')
              .merge(long_tot, on='Act Team', how='left')
              .merge(long_succ, on='Act Team', how='left')
              .merge(in_att_3rd_tot, on='Act Team', how='left')
              .merge(in_att_3rd_succ, on='Act Team', how='left')
              .merge(into_att_3rd_tot, on='Act Team', how='left')
              .merge(into_att_3rd_succ, on='Act Team', how='left')
              .merge(cross_tot, on='Act Team', how='left')           
              .merge(cross_succ, on='Act Team', how='left')
              .merge(missed, on='Act Team', how='left')              # <-- Added Specific Fails
              .merge(intercepted, on='Act Team', how='left')         # <-- Added Specific Fails
              .merge(offside, on='Act Team', how='left')             # <-- Added Specific Fails
              .merge(prog_tot, on='Act Team', how='left')            # <-- Added Progressive
              .merge(prog_succ, on='Act Team', how='left')           # <-- Added Progressive
              .merge(prog_opp_tot, on='Act Team', how='left')        # <-- Added Progressive Opp Half
              .merge(prog_opp_succ, on='Act Team', how='left')
              .merge(into_box_tot, on='Act Team', how='left')        # <-- Added
              .merge(into_box_succ, on='Act Team', how='left')       # <-- Added
              .merge(inside_box_tot, on='Act Team', how='left')      # <-- Added
              .merge(inside_box_succ, on='Act Team', how='left'))     # <-- Added)      # <-- Added Progressive Opp Half

# 15. Fill any missing values with 0
pass_stats = pass_stats.fillna(0)

# 16. Calculate Overall Pass Accuracy
pass_stats['Pass Accuracy'] = round((pass_stats['Complete Pass'] / pass_stats['Total Pass']) * 100, 0)

pass_stats
```

```python
import numpy as np

# 1. First, make sure you have calculated Distance and Angle
shots = shoot_df.copy()
shots.loc[shots['Type'] == 'Free Kick', 'Pressure'] = 'No'
shots.loc[shots['Type'] == 'Free Kick', 'Assist Type'] = 'Normal Pass'

shots.loc[shots['Minute'] == 50, 'Pressure'] = 'No'
shots.loc[shots['Minute'] == 46, 'Pressure'] = 'No'
shots.loc[shots['Minute'] == 41, 'Pressure'] = 'No'
shots['Distance'] = np.sqrt((120 - shots['Start X'])**2 + (40 - shots['Start Y'])**2)

def calc_angle(x, y):
    v1_x, v1_y = 120 - x, 36 - y
    v2_x, v2_y = 120 - x, 44 - y
    dot = v1_x*v2_x + v1_y*v2_y
    mag1 = np.sqrt(v1_x**2 + v1_y**2)
    mag2 = np.sqrt(v2_x**2 + v2_y**2)
    cos_angle = np.clip(dot / (mag1 * mag2 + 1e-9), -1.0, 1.0)
    return np.degrees(np.arccos(cos_angle))

shots['Angle'] = shots.apply(lambda row: calc_angle(row['Start X'], row['Start Y']), axis=1)

# 2. The Custom xG Formula
def calculate_xg(row):
    # Penalties are a fixed probability globally (~76%)
    if row['Type'] == 'Penalty':
        return 0.76
        
    # RULE 1: Extreme long shots (own half / far out) get max 1% probability
    if row['Start X'] <= 80:
        return 0.01
        
    # --- TUNE THESE COEFFICIENTS ---
    z = -0.5 
    
    # CORRECTED: Direct Free Kicks have a wall in the way
    if row['Type'] == 'Free Kick':
        z += 0.7
    
    # Geometry 
    z -= 0.10 * row['Distance']   
    z += 0.012 * row['Angle']     
    
    # Shot Technique
    if row['Shot Technique'] == 'Header':
        z -= 0.80  
    elif row['Shot Technique'] == 'Volley':
        z -= 0.2  
        
    # Assist Type
    if row['Assist Type'] == 'Through ball':
        z += 0.6  
    elif row['Assist Type'] == 'Cross':
        z -= 0.4  
    elif row['Assist Type'] == 'Rebound':
        z += 0.5  
    elif row['Assist Type'] == 'Free Kick':
        z += 0.3   
        
    # Context
    if row['First Time Shot'] == 1:
        z += 0.4  
        
    # --- PRESSURE CHECKS ---
    if row['Pressure'] == 'Yes':
        z -= 0.90  
    elif row['Pressure'] == 'No' and row['Outcome'] == 'Block':
        # NEW RULE: If there was no recorded pressure but it was blocked, 
        # there was clearly a defender in the shot path.
        z -= 0.60
        
    # Calculate final probability using the Sigmoid curve
    xg = 1 / (1 + np.exp(-z))
    
    # RULE 2: If xG drops below 0.01 anywhere inside the attacking half (>80), bump it to 0.02
    if row['Start X'] > 80 and xg < 0.01:
        xg = 0.02
        
    # RULE 3: New Danger Zone Floor (Edge of box / central areas)
    if (95 <= row['Start X'] <= 103) and (15 <= row['Start Y'] <= 65) and xg < 0.02:
        xg = 0.02
        
    return xg


# 3. The Custom xGOT Formula
def calculate_xgot(row):
    # If the shot missed or hit the woodwork, xGOT is strictly 0
    if row['Outcome'] not in ['Goal', 'Save']:
        return 0.0
        
    base_xg = row['xG']
    placement_y = abs(row['End Y'] - 40)
    
    # --- TUNE THESE MULTIPLIERS ---
    if placement_y >= 3.0:
        multiplier = 1.6 
    elif placement_y >= 1.5:
        multiplier = 1.1 
    else:
        multiplier = 0.6 
        
    # Apply multiplier and cap it at 99%
    xgot = min(base_xg * multiplier, 0.99)
    return xgot


# Apply the functions to your dataset
shots['xG'] = shots.apply(calculate_xg, axis=1)
shots['xGOT'] = shots.apply(calculate_xgot, axis=1)
```

```python
import pandas as pd
import numpy as np

# 1. Filter for the two sides of the event
intended_passes = df[(df.Action == 'Pass') & (df.Outcome == 'Intercepted')]
header_intercepts = df[(df.Action == 'Pass Intercept') & (df.Outcome == 'Successful') & (df['Body Part'] == 'Head')]

# 2. Thresholds
dist_threshold = 3.0
time_threshold = 8.0 

matches = []

for idx, interception in header_intercepts.iterrows():
    # Flip defender coordinates to Attacker Space
    def_x_flipped = 120 - interception['Start X']
    def_y_flipped = 80 - interception['Start Y']
    
    # 3. Find passes within time window
    potential_pass = intended_passes[
        (abs(intended_passes['Time (s)'] - interception['Time (s)']) <= time_threshold)
    ].copy()
    
    if not potential_pass.empty:
        # Calculate distance for all potential candidates
        potential_pass['calculated_dist'] = np.sqrt(
            (potential_pass['End X'] - def_x_flipped)**2 + 
            (potential_pass['End Y'] - def_y_flipped)**2
        )
        
        # Filter by distance threshold
        actual_matches = potential_pass[potential_pass['calculated_dist'] <= dist_threshold]
        
        if not actual_matches.empty:
            # Take the one with the smallest distance (the most likely duel)
            best_match = actual_matches.sort_values('calculated_dist').iloc[0]
            
            matches.append({
                'Time': interception['Time (s)'],
                'Defender': interception['Act Player'],
                'Attacker_Intended': best_match['React Player'],
                'X_Def_Raw': interception['Start X'], # Original defender coordinate
                'Y_Def_Raw': interception['Start Y'],
                'X_Att_End': best_match['End X'],     # Intended target coordinate
                'Y_Att_End': best_match['End Y'],
                'Distance': best_match['calculated_dist'] # The distance you asked for
            })

aerial_duels_df = pd.DataFrame(matches)
aerial_duels_df['Min']=aerial_duels_df['Time']/60
aerial_duels_df
```

```python
import pandas as pd
import numpy as np

# 1. Filter for the two sides of the event
intended_passes = df[(df['Action'] == 'Pass') & (df['Outcome'] == 'Intercepted')]
header_intercepts = df[(df['Action'] == 'Pass Intercept') & (df['Outcome'] == 'Successful') & (df['Body Part'] == 'Head')]

# 2. Thresholds
dist_threshold = 3.0
time_threshold = 8.0 

matches = []

for idx, interception in header_intercepts.iterrows():
    # Flip defender coordinates to Attacker Space
    def_x_flipped = 120 - interception['Start X']
    def_y_flipped = 80 - interception['Start Y']
    
    # 3. Find passes within time window
    potential_pass = intended_passes[
        (abs(intended_passes['Time (s)'] - interception['Time (s)']) <= time_threshold)
    ].copy()
    
    if not potential_pass.empty:
        # Calculate distance for all potential candidates
        potential_pass['calculated_dist'] = np.sqrt(
            (potential_pass['End X'] - def_x_flipped)**2 + 
            (potential_pass['End Y'] - def_y_flipped)**2
        )
        
        # Filter by distance threshold
        actual_matches = potential_pass[potential_pass['calculated_dist'] <= dist_threshold]
        
        if not actual_matches.empty:
            # Take the one with the smallest distance
            best_match = actual_matches.sort_values('calculated_dist').iloc[0]
            
            matches.append({
                'Time': interception['Time (s)'],
                'Act Team': interception['Act Team'], # <-- ADDED THIS to allow aggregation
                'Defender': interception['Act Player'],
                'Attacker_Intended': best_match['React Player'],
                'X_Def_Raw': interception['Start X'], 
                'Y_Def_Raw': interception['Start Y'],
                'X_Att_End': best_match['End X'],     
                'Y_Att_End': best_match['End Y'],
                'Distance': best_match['calculated_dist'] 
            })

# Create the Duels DataFrame
aerial_duels_df = pd.DataFrame(matches)

# Safety check: Only calculate 'Min' if the dataframe isn't empty
if not aerial_duels_df.empty:
    aerial_duels_df['Min'] = aerial_duels_df['Time'] / 60
```

```python
# 1. Aggregate the successful aerial duels by team
if not aerial_duels_df.empty:
    total_aerial_duels = aerial_duels_df['Act Team'].value_counts()
else:
    # Fallback in case no matches were found to prevent errors
    total_aerial_duels = pd.Series(dtype=int)

# 2. Add as a new column to the main table
team_defensive_stats['Aerial Duels Won'] = total_aerial_duels

# 3. Fill any NaNs with 0 (for teams that didn't win any aerial duels) and cast to integer
team_defensive_stats['Aerial Duels Won'] = team_defensive_stats['Aerial Duels Won'].fillna(0).astype(int)

# Display the final updated table
team_defensive_stats
```

```python
# 1. Define successful pass outcomes
success_outcomes = ['Successful', 'Key Pass', 'Assist']

# 2. Passes Allowed (The opponent passing in your high-press zone)
# Fix: Added ['Action'] to the first condition
ppda_pass = df[
    (df['Action'].isin(['Pass', 'Through Ball'])) & 
    (df['Start X'] < 72) & 
    (df['Outcome'].isin(success_outcomes)) & 
    (df['Type'].isin(['Normal', 'Normal Pass']))
].copy()

# 3. Defensive Actions (Your team pressing in that same zone)
defensive_actions = ['Block', 'Pass Intercept', 'Standing Tackle', 'Sliding Tackle']

ppda_def = df[
    (df['Action'].isin(defensive_actions)) & 
    (df['Start X'] > 48)
].copy()

# 1. Count the passes made by each team in their own defensive zone
passes_made = ppda_pass['Act Team'].value_counts().reset_index(name='Passes Made')

# 2. Count the defensive actions made by each team in their high-press zone
def_actions = ppda_def['Act Team'].value_counts().reset_index(name='Def Actions')

# 3. Merge them together
ppda_calc = passes_made.merge(def_actions, on='Act Team', how='outer').fillna(0)

# 4. The "Opponent Swap" Trick
# We need to shift the passes so Team A's passes are applied to Team B's formula.
# We reverse the 'Passes Made' column and assign it to a new 'Passes Allowed' column
ppda_calc['Passes Allowed'] = ppda_calc['Passes Made'].iloc[::-1].reset_index(drop=True)

# 5. Calculate Final PPDA! (Passes Allowed / Defensive Actions)
ppda_calc['PPDA'] = round(ppda_calc['Passes Allowed'] / ppda_calc['Def Actions'], 2)

# Display the final, clean table (dropping the 'Passes Made' column to avoid confusion)
final_ppda = ppda_calc[['Act Team', 'Passes Allowed', 'Def Actions', 'PPDA']]
final_ppda
```

```python
import os
import shutil
from pathlib import Path

def extract_all_images(source_folder, destination_folder):
    """
    Finds all images in the source_folder and its subfolders, 
    then copies them to the destination_folder.
    """
    # 1. Create the destination folder if it doesn't exist
    dest_path = Path(destination_folder)
    dest_path.mkdir(parents=True, exist_ok=True)
    
    source_path = Path(source_folder)
    
    # 2. Define what file extensions count as images
    image_extensions = {'.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff', '.webp'}
    
    copied_count = 0
    print(f"Scanning '{source_folder}' for images...")

    # 3. Search through all folders and subfolders
    for file_path in source_path.rglob('*'):
        # Check if it's a file and has an image extension
        if file_path.is_file() and file_path.suffix.lower() in image_extensions:
            
            # 4. Handle duplicate file names so nothing is overwritten
            dest_file = dest_path / file_path.name
            counter = 1
            while dest_file.exists():
                # If 'image.jpg' exists, try 'image_1.jpg', then 'image_2.jpg', etc.
                new_name = f"{file_path.stem}_{counter}{file_path.suffix}"
                dest_file = dest_path / new_name
                counter += 1
            
            # 5. Copy the file
            try:
                shutil.copy2(file_path, dest_file)
                copied_count += 1
            except Exception as e:
                print(f"Warning: Could not copy {file_path.name}. Error: {e}")

    print(f"\nSuccess! Copied {copied_count} images to '{destination_folder}'.")

# ==========================================
# HOW TO USE:
# Replace the paths below with your actual folder paths.
# Keep the 'r' before the quotes if you are on Windows, or remove it for Mac paths.
# ==========================================

# Example for Mac:
MY_MAIN_FOLDER = '/Users/preciousprince/Downloads/KabaddiImages'
MY_NEW_FOLDER = '/Users/preciousprince/Desktop/kabaddi'

extract_all_images(MY_MAIN_FOLDER, MY_NEW_FOLDER)
```
